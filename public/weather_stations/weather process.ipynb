```python
import os
import pandas as pd
"""
这一段是获取每一天 每小时的值s
"""

# 获取文件夹中的所有文件
folder_path = 'C:/Users/SuYunlei/Documents/VC_data_csv'
file_names = os.listdir(folder_path)

# 创建一个空的DataFrame来存储结果
result_df = pd.DataFrame()

# 循环处理每个文件
for file_name in file_names:
    if file_name.endswith('.csv'):  # 确保只处理 CSV 文件
        file_path = os.path.join(folder_path, file_name)

        # 读取CSV文件并解析日期和时间列
        df = pd.read_csv(file_path, parse_dates={'DateTime': ['Date', 'Time']})

        # 设置DateTime列为索引
        df.set_index('DateTime', inplace=True)

        # 按小时计算每列的平均值
        hourly_mean = df.resample('H').mean()

        # 将当前文件的数据添加到结果DataFrame中
        result_df = pd.concat([result_df, hourly_mean])

# 清理缺失数据的列
# result_df_cleaned = result_df.dropna(axis=1)

# 保存为新的CSV文件
result_df.to_csv('hourly_averages.csv')
```

```python
"""
这一段是获取每一个月 每小时的值
"""
# 读取CSV文件
data = pd.read_csv('hourly_averages.csv')

# 将DateTime列转换为日期时间类型
data['DateTime'] = pd.to_datetime(data['DateTime'])

# 提取月份和小时
data['Month'] = data['DateTime'].dt.month
data['Hour'] = data['DateTime'].dt.hour

# 计算相同月份、不同小时的平均值
average_data = data.groupby(['Month', 'Hour']).mean().reset_index()

# 清理缺失数据的列
# avarage_data_cleaned = average_data.dropna(axis=1)

# 保存为新的CSV文件
average_data.to_csv('average_data.csv', index=False)
```

```python
"""
这一段本该是按照传感器分离数据的...
"""

# 读取CSV文件
avg_data = pd.read_csv('average_data.csv')

# 获取传感器列名
header = avg_data.columns.tolist()

array = [[] for _ in range(21)]

for column in header:
    for i in range(1, 21):
        if f'#0{i}' in column:
            array[i].append(column)

for column in header:
    for i in range(10, 21):
        if f'#{i}' in column:
            array[i].append(column)
            
array = array[1:]

for index,ele in enumerate(array):
    temp = ['Month','Hour']
    for e in ele:
        temp.append(e)
    selected_data = avg_data[temp]
    file_name = f'weather{index+1}.csv'
    selected_data.to_csv(file_name)
    print(index)
```
